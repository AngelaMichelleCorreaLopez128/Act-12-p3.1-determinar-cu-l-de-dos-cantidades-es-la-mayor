```python
# calcula la diferencia de valor de if else
# datos de entrada A,B
# mostrar B si es falso
# mostrar A si es verdadero
A = int(input("Dame el primer valor: "))
B = int(input("Dame el segundo valor: "))
if A > B:
  M = A
else:
  M = B
print("El mayor valor es: ", M)
print("fin del programa")
print("programa realizado por Angela Michelle Correa Lopez NL = 13")
```
